```python
hospitals=pd.read_csv('./covid19/Covid_data_srk/ICMRTestingLabs.csv')
hospitals.rename(columns={"pincode": "Pincode"},inplace=True)
hospitals_mumbai=hospitals.loc[hospitals['city']== 'Mumbai']
hospitals_mumbai
```

```python
hospitals_mumbai
```

```python
mumbai_labs_address=hospitals_mumbai.merge(mumbai_pincodes,how='left',on='Pincode',validate='m:1')
```

```python
mumbai_labs_address
```

```python
mumbai_covid_labs=mumbai_labs_address[['lab','address','Pincode','Latitude','Longitude','type']]
```

```python
type(mumbai_covid_labs)
```

```python
#Plot Toronto city on Map using folium
mumbai= folium.Map(location=[latitude, longitude], zoom_start=9)
for lat, lng, Pincode, lab,typeh in zip(mumbai_covid_labs['Latitude'], mumbai_covid_labs['Longitude'], mumbai_covid_labs['Pincode'], mumbai_covid_labs['lab'],mumbai_covid_labs['type']):
    label = '{}, {},{}'.format(Pincode,lab,typeh)
    label = folium.Popup(label, parse_html=True)
    if typeh =='Private Laboratory':
        val='blue'
    else:
        val='green'
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=val,
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(mumbai)  

mumbai
```
