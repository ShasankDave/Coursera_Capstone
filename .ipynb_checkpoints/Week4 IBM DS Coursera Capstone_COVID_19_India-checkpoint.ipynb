```python
#importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
covid_19_ind=pd.read_csv('/mnt/d/Coursera_Capstone/covid19/Covid_data_srk/covid_19_india.csv')
data = covid_19_ind.copy()
data['Date'] = data['Date'].apply(pd.to_datetime)
data.drop(['Sno', 'Time'],axis=1,inplace=True)

# collect present data
from datetime import date
data_apr = data[data['Date'] > pd.Timestamp(date(2020,4,23))]

# prepaing data state wise
state_cases = data_apr.groupby('Date')['Confirmed','Deaths','Cured'].sum().reset_index()
#state_cases['Active'] = state_cases['Confirmed'] - (state_cases['Deaths']- state_cases['Cured'])
#state_cases["Death Rate (per 100)"] = np.round(100*state_cases["Deaths"]/state_cases["Confirmed"],2)
state_cases
```

```python
state_cases.plot(x='Date',y=['Confirmed','Deaths','Cured'],kind='bar')
plt.show()
```
